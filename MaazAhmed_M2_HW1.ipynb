###### 4. Checking Null values and handling them

```python
data.isnull().sum()
```

###### 4.1 Droping all the columns where NULL values are more then 70%

```python
data.dropna(axis=1, thresh=0.70*len(data), inplace=True)
```

```python
data.shape
```

###### 4.2 Droping all the rows where NULL values in all columns


```python
data.dropna(how='all',subset=["ConsecutiveWins","Season","Opposition_Others","Country","Ground","Home_Ground","Day_Night","PitchReport","BattingFirst","Pak_Win"])
```

###### 4.3 Droping all the row where NULL values are more then 70%

```python
# Removeing rows having 70% of NAN values .....
# havent save cz i want to fill that data..

data.dropna(axis=0, thresh=0.70*len(data.columns), inplace=True)
```

```python
data.isnull().sum()
```

```python
# filling missing values

# for season

for i, row in enumerate(data['Season']):
    if row is np.nan:
        print(data['Season'].iloc[i-5:i+5].mode())
        print(data.Season.iloc[i])
        data.Season.iloc[i]=data['Season'].iloc[i-5:i+5].mode()[0]
#         data.iloc[i,row]=data['Season'].iloc[i-5:i+5].mode()[]
    else:
        pass
```

```python
# for Opposition_Others

for i, row in enumerate(data['Opposition_Others']):
    if row is np.nan:
        data.Opposition_Others.iloc[i] = data.Opposition_Others.loc[data['Country'] == data.loc[i,'Country']].mode()[0]
#         data.Season.iloc[i]=data['Season'].iloc[i-5:i+5].mode()[0]
#         data.iloc[i,row]=data['Season'].iloc[i-5:i+5].mode()[]
    else:
        pass
```

###### 5. Dealing with incorrect values and changing datatypes

```python
# for ConsecutiveWins

for i , row in enumerate(data['ConsecutiveWins']):
    if(row.isnumeric()):
        pass
    else:
        print(i , row)
        data.loc[i,'ConsecutiveWins']= data['ConsecutiveWins'].mode()[0]
```

```python
data['ConsecutiveWins']=data['ConsecutiveWins'].astype(int)
```

```python
# for Home_Ground

for i , row in enumerate(data['Home_Ground']):
    if(row == 'Yes' or row == 'yes'):
        data.loc[i,'Home_Ground']= '1'
    else:
        data.loc[i,'Home_Ground']= '0'
#         data.loc[i,'Home_Ground']= data['Home_Ground'].mode()[0]
```

```python
data['Home_Ground']=data['Home_Ground'].astype(int)
```

```python
# for Day_Night

for i , row in enumerate(data['Day_Night']):
    if(row == 'Yes' or row == 'yes'):
        data.loc[i,'Day_Night']= '1'
    else:
        data.loc[i,'Day_Night']= '0'
#         data.loc[i,'Home_Ground']= data['Home_Ground'].mode()[0]
```

```python
data['Day_Night']=data['Day_Night'].astype(int)
```

```python
# for BattingFirst

for i , row in enumerate(data['BattingFirst']):
    if(row == 'Yes' or row == 'yes'):
        data.loc[i,'BattingFirst']= '1'
    else:
        data.loc[i,'BattingFirst']= '0'
#         data.loc[i,'Home_Ground']= data['Home_Ground'].mode()[0]
```

```python
data['BattingFirst']=data['BattingFirst'].astype(int)
```

```python
# for Pak_Win

for i , row in enumerate(data['Pak_Win']):
    if(row == 'Yes' or row == 'yes'):
        print(row)
        data.loc[i,'Pak_Win']= '1'
    elif (row == 'No' or row == 'No'):
        print(i , row)
        data.loc[i,'Pak_Win']= '0'
    else:
        data.loc[i,'Pak_Win']= data['Pak_Win'].mode()[0]
```

```python
data['Pak_Win']=data['Pak_Win'].astype(int)
```
